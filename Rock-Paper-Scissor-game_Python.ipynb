```python
import tkinter
import random
import tkinter.messagebox

# Create Object 
root = tkinter.Tk() 
 
# Set geometry 
root.geometry("400x500") 
 
# Set title 
root.title("Rock-Paper-Scissors Game")

# Computer Value 
computer_dict = { 
    "0": "Rock", 
    "1": "Paper", 
    "2": "Scissors" 
}

# Number of rounds and scores
rounds = 0
current_round = 0
user_score = 0
total_score = 0

# Function to enable/disable buttons
def toggle_buttons(state):
    b1.config(state=state)
    b2.config(state=state)
    b3.config(state=state)

# Function to start the game
def start_game():
    global rounds, current_round, user_score, total_score
    rounds_input = entry.get()
    if rounds_input.isdigit() and int(rounds_input) > 0:  # Check if input is a positive integer
        rounds = int(rounds_input)
        current_round = 0
        user_score = 0
        total_score = 0
        entry.config(state="disabled")
        play_button.config(state="disabled")
        toggle_buttons("normal")  # Enable buttons
        l1.config(text="Player             ") 
        l3.config(text="Computer") 
        update_rounds()
        update_scores()
    else:
        tkinter.messagebox.showerror("Error", "Please enter a valid number of rounds.")

# Function to reset the game 
def reset_game(): 
    global rounds, current_round, user_score, total_score
    entry.config(state="normal")
    play_button.config(state="normal")
    toggle_buttons("disabled")  # Disable buttons
    l1.config(text="Player             ") 
    l3.config(text="Computer")
    last_result = l4.cget("text")
    if last_result:
        l4.config(text="Last Round: " + last_result)  # Displaying last round result
        if current_round == rounds:
            total_score += user_score  # Add last round's user score to total score
    else:
        l4.config(text="Last Round: No rounds played")
    rounds = 0
    current_round = 0
    user_score = 0
    entry.delete(0, tkinter.END)
    update_rounds()
    update_scores()

# Update number of rounds label
def update_rounds():
    rounds_label.config(text=f"Rounds: {current_round}/{rounds}")

# Update user score label
def update_scores():
    user_score_label.config(text=f"User Score: {user_score}/{current_round}")
    total_score_label.config(text=f"Total Score: {total_score}/{current_round}")

# If player selects rock 
def player_rock(): 
    global current_round, user_score, total_score
    c_v = computer_dict[str(random.randint(0,2))] 
    if c_v == "Rock": 
        match_result = "Tie!" 
    elif c_v == "Scissors": 
        match_result = "Player Wins" 
        user_score += 1
    else: 
        match_result = "Computer Wins" 
        total_score += 1
    l4.config(text = match_result) 
    l1.config(text = "Rock         ") 
    l3.config(text = c_v) 
    current_round += 1
    update_rounds()
    update_scores()
    toggle_buttons("disabled")  # Disable buttons
    if current_round == rounds:
        reset_game()
    else:
        toggle_buttons("normal")  # Enable buttons for the next round

# If player selects paper 
def player_paper(): 
    global current_round, user_score, total_score
    c_v = computer_dict[str(random.randint(0,2))] 
    if c_v == "Paper": 
        match_result = "Tie!" 
    elif c_v == "Scissors": 
        match_result = "Computer Wins" 
        total_score += 1
    else: 
        match_result = "Player Wins" 
        user_score += 1
    l4.config(text = match_result) 
    l1.config(text = "Paper        ") 
    l3.config(text = c_v) 
    current_round += 1
    update_rounds()
    update_scores()
    toggle_buttons("disabled")  # Disable buttons
    if current_round == rounds:
        reset_game()
    else:
        toggle_buttons("normal")  # Enable buttons for the next round

# If player selects scissors 
def player_scissors(): 
    global current_round, user_score, total_score
    c_v = computer_dict[str(random.randint(0,2))] 
    if c_v == "Rock": 
        match_result = "Computer Wins" 
        total_score += 1
    elif c_v == "Scissors": 
        match_result = "Tie!" 
    else: 
        match_result = "Player Wins" 
        user_score += 1
    l4.config(text = match_result) 
    l1.config(text = "Scissors         ") 
    l3.config(text = c_v) 
    current_round += 1
    update_rounds()
    update_scores()
    toggle_buttons("disabled")  # Disable buttons
    if current_round == rounds:
        reset_game()
    else:
        toggle_buttons("normal")  # Enable buttons for the next round

# Add Labels, Frames and Button 
tkinter.Label(root, 
    text = 'Choose any one: Rock, Paper, Scissors', 
    font = "Consolas", 
    fg = "blue").pack(pady = 20)  


frame = tkinter.Frame(root) 
frame.pack() 

l1 = tkinter.Label(frame, 
    text = "Player         ", 
    font = 10) 

l2 = tkinter.Label(frame, 
    text = "VS         ", 
    font = "Consolas") 

l3 = tkinter.Label(frame, text = "Computer", font = 10) 

l1.pack(side = 'left') 
l2.pack(side = 'left') 
l3.pack() 

l4 = tkinter.Label(root, 
        text = "", 
        font = "Consolas", 
        bg = "white", 
        width = 20 , 
        borderwidth = 2, 
        relief = "solid") 
l4.pack(pady = 30) 

frame1 = tkinter.Frame(root) 
frame1.pack() 

b1 = tkinter.Button(frame1, text = "Rock", 
            font = 8, width = 7, bg = "light blue", 
            command = player_rock,
            state="disabled") 

b2 = tkinter.Button(frame1, text = "Paper ", 
            font = 8, width = 7, bg = "light blue", 
            command = player_paper,
            state="disabled") 

b3 = tkinter.Button(frame1, text = "Scissors", 
            font = 8, width = 7, bg = "light blue", 
            command = player_scissors,
            state="disabled") 

b1.pack(side = 'left', padx = 10) 
b2.pack(side = 'left', padx = 10) 
b3.pack(padx = 10) 

entry_frame = tkinter.Frame(root)
entry_frame.pack(pady=10)
entry_label = tkinter.Label(entry_frame, text="Enter number of rounds: ")
entry_label.pack(side="left")
entry = tkinter.Entry(entry_frame, width=5)
entry.pack(side="left")

play_button = tkinter.Button(root, text="Start Game", font=10, fg="green", bg="light grey", command=start_game)
play_button.pack(pady=10)

rounds_label = tkinter.Label(root, text="Rounds: 0/0", font="Consolas")
rounds_label.pack()

user_score_label = tkinter.Label(root, text="User Score: 0/0", font="Consolas")
user_score_label.pack()

total_score_label = tkinter.Label(root, text="Total Score: 0/0", font="Consolas")  # Corrected
total_score_label.pack()

tkinter.Button(root, text = "Reset Game", 
        font = 10, fg = "red", 
        bg = "light grey", command = reset_game).pack(pady = 20) 

# Execute Tkinter 
root.mainloop()
```
